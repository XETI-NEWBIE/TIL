### 핵심 개념
- **청킹 전략(Chunking Strategy)**: chunk_size(청크 크기)와 overlap(겹침)이 검색 결과에 미치는 영향. 작은 청크는 세밀한 검색이 가능하나 문맥 부족, 큰 청크는 문맥 풍부하나 노이즈 포함 위험
- **Retriever 파라미터**: top_k(반환 문서 수) 등 검색 설정이 RAG 응답 품질에 미치는 영향. 적은 k는 정보 누락 위험, 많은 k는 관련 없는 정보 포함 위험
- **설정 비교/분석**: 실험 결과를 바탕으로 상황에 맞는 최적 설정을 판단하고 근거를 제시하는 역량

# Step 2: 청킹 전략 변경 및 비교

### Concept Check: chunk_size와 overlap의 영향

| 설정 | 작은 값 | 큰 값 |
|------|---------|-------|
| chunk_size | 세밀한 검색, 문맥 부족 | 넓은 문맥, 검색 정확도 하락 |
| overlap | 문맥 단절 위험 | 중복 정보, 비용 증가 |

**실험 목표**: 다양한 설정으로 검색 품질이 어떻게 달라지는지 비교

### Guided Build: 청킹 및 Vector Store 생성 함수

아래 함수를 사용하여 다양한 설정으로 실험합니다.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
import tiktoken

def create_vectorstore_with_chunking(documents, chunk_size, chunk_overlap, collection_name):
    """
    주어진 청킹 설정으로 Vector Store를 생성합니다.
    
    Args:
        documents: 원본 문서 리스트
        chunk_size: 청크 크기
        chunk_overlap: 청크 간 겹침
        collection_name: ChromaDB 컬렉션 이름
    
    Returns:
        vectorstore, chunked_documents
    """
    # 청킹
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap,
        length_function=len,
        separators=["\n\n", "\n", ".", " ", ""]
    )
    chunked_docs = text_splitter.split_documents(documents)
    
    # 임베딩 모델의 토큰 제한 안전장치 (text-embedding-3-small: 최대 8,191 토큰)
    MAX_TOKENS = 8000
    enc = tiktoken.encoding_for_model("text-embedding-3-small")
    for doc in chunked_docs:
        tokens = enc.encode(doc.page_content)
        if len(tokens) > MAX_TOKENS:
            doc.page_content = enc.decode(tokens[:MAX_TOKENS])
    
    # Vector Store 생성
    vectorstore = Chroma.from_documents(
        documents=chunked_docs,
        embedding=embeddings,
        collection_name=collection_name
    )
    
    return vectorstore, chunked_docs

print("✅ 함수 정의 완료")
```

# Step 3: Retriever 설정 변경 및 비교

### Concept Check: top_k의 영향

**top_k**: 검색 시 반환할 문서 개수

| top_k 값 | 특징 |
|----------|------|
| 작음 (1~2) | 가장 관련성 높은 문서만, 정보 누락 위험 |
| 중간 (3~5) | 균형 잡힌 검색 |
| 큼 (5+) | 다양한 정보, 관련 없는 내용 포함 위험 |

### Guided Build: RAG 체인 생성 함수

아래 함수를 사용하여 다양한 top_k로 RAG 결과를 비교합니다.

```python
from langchain_core.prompts import ChatPromptTemplate
# 기본 설정으로 Vector Store 생성 (비교 실험용)
vectorstore_base, _ = create_vectorstore_with_chunking(
    documents=all_documents,
    chunk_size=300,
    chunk_overlap=50,
    collection_name="yes24_base"
)

# RAG 프롬프트
rag_prompt = ChatPromptTemplate.from_messages([
    ("system", """당신은 온라인 서점 'Yes24'의 고객 서비스 상담원입니다.
다음 검색된 자료를 참고하여 고객의 질문에 정확하고 친절하게 답변해주세요.

[검색된 자료]
{context}

답변 규칙:
1. 검색된 자료를 기반으로 답변하세요
2. 자료에 없는 내용은 추측하지 마세요"""),
    ("human", "{question}")
])

def format_docs(docs):
    return "\n\n".join([doc.page_content for doc in docs])

def run_rag(retriever, question):
    """
    주어진 retriever와 질문으로 RAG를 수행합니다.
    """
    # 관련 문서 검색
    docs = retriever.invoke(question)
    context = format_docs(docs)
    
    # 프롬프트 생성 후 LLM 직접 호출
    messages = rag_prompt.format_messages(context=context, question=question)
    response = llm.invoke(messages)
    
    return response.content

def create_rag_components(vectorstore, k):
    """
    주어진 top_k 설정으로 RAG 컴포넌트를 생성합니다.
    """
    retriever = vectorstore.as_retriever(search_kwargs={"k": k})
    return retriever

print("✅ 함수 정의 완료")
```

#### TODO 4 답안: top_k 차이점 분석

**관찰 결과:**
- top_k=2: 가장 질문과 똑같은 핵심 문서 딱 2개만 가져옵니다. 답변은 군더더기 없이 간결하지만, 복합적인 질문(예: 반품도 하고 포인트도 쓰고 싶어요)에는 정보가 모자랄 수 있습니다.
- top_k=5: 관련된 문서를 무려 5개나 긁어오므로 정보는 엄청 풍부해집니다. 하지만 질문과 덜 관련된 엉뚱한 규정이 섞여 들어가서 챗봇이 할루시네이션을 할 위험이 있습니다.

**결론**: 무조건 문서를 많이 가져온다고 좋은 것이 아니며, 정보 누락과 노이즈 발생 사이의 적절한 타협점(Trade-off)을 찾아야 합니다.

##### 선택한 설정:

chunk_size: 300

##### 선택 근거:

chunk_size 선택 이유: Yes24 규정 안내문은 비교적 간결하므로, 500은 너무 커서 노이즈가 끼고 200은 문맥이 잘릴 위험이 있어 중간값인 300이 적당합니다.

chunk_overlap 선택 이유: 문장이 끊어질 때 정보가 이어지도록 50 정도의 여유를 두어 문맥 유실을 방지했습니다.

top_k 선택 이유: 2개는 복합 질문 방어가 어렵고 5개는 오답 확률이 높으므로, 3개의 문서를 참고하는 것이 응답 정확도와 풍부함의 밸런스가 가장 좋습니다.

```python
# 최종 설정으로 RAG 파이프라인 구성
# TODO: 위에서 선택한 설정으로 아래 값들을 수정하세요

FINAL_CHUNK_SIZE = 300      # 1. 선택한 값으로 수정
FINAL_CHUNK_OVERLAP = 50    # 2. 선택한 값으로 수정
FINAL_TOP_K = 3             # 3. 선택한 값으로 수정

final_vectorstore, final_chunks = create_vectorstore_with_chunking(
    documents=all_documents,
    chunk_size=FINAL_CHUNK_SIZE,
    chunk_overlap=FINAL_CHUNK_OVERLAP,
    collection_name="yes24_final"
)

final_retriever = create_rag_components(final_vectorstore, k=FINAL_TOP_K)

print("✅ 최종 RAG 파이프라인 구성 완료")
print(f"📊 설정: chunk_size={FINAL_CHUNK_SIZE}, overlap={FINAL_CHUNK_OVERLAP}, top_k={FINAL_TOP_K}")
print(f"📊 총 청크 수: {len(final_chunks)}개")
```

```python
# 최종 파이프라인 테스트
test_questions = [
    "총알배송이 뭔가요?",
    "배송이 늦으면 보상받을 수 있나요?",
    "신규 회원 혜택은 무엇인가요?"
]

print("🧪 최종 RAG 파이프라인 테스트")
print("="*60)

for q in test_questions:
    print(f"\n📝 질문: {q}")
    print("\n🤖 응답:")
    response = run_rag(final_retriever, q)
    print(response)
    print("-"*60)
```
