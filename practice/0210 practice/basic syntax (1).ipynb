#### 소수점 형식지정
f-string 안에서 소수점 n번째 자리까지 작성한다.
.n 을 통해 원하는 자릿수까지 출력이 가능하다. f는 float 을 의미한다.
print(f'{  }') 안에서 { } 안에 :.nf 형식으로 작성한다 

: 소수점 이하 자릿수는 {변수:포맷} 형식을 사용하여 :.n 과 같은 구문을 통해 원하는 자릿수까지 출력할 수 있다.
n 지정시 소수점 n+1 번째 자리에서 반올림하여 소수점 n 번째 자리까지 출력한다.

pi = 3.141592653589793

#### 소수점 이하 2자리까지 반올림하여 출력

print(f"파이: {pi:.2f}")

출력: 파이: 3.14

=> pi 를 소수점 2번째 자리까지 반올림하여 출력

#### 소수점 이하 4자리까지 반올림하여 출력

print(f"파이: {pi:.4f}")

출력: 파이: 3.1416

=> 두 번째 예제에선 소수점 4번째 자리까지 반올림하여 출력

## end=", sep="" 파라미터

기본적으로 print() 함수는 출력 후 줄바꿈을 수행하여 다음 출력이 새로운 줄에 나타난다.
print 함수 두 줄에 연속으로 나온다면, end 파라미터는 print()를 이어주는 역할을 함

```python
ex) 기본 print() 출력

print("첫 번째 줄")
print("두 번째 줄")

 출력:
첫 번째 줄
두 번째 줄
```

### end= 파라미터
- print () 함수의 출력이 끝난 후 추가할 문자열을 지정
- 기본값은 줄바꿈 문자(\n)로, 출력 후 자동 줄바꿈이 수행된다.

#### 1. 줄바꿈 대신 공백으로 변경
- end를 공백("")으로 설정하면, 두 출력이 같은 줄에 공백으로 구분되어 이어서 출력된다.
- print() 호출 후 공백을 추가하면 모든 출력이 같은 줄에 공백으로 구분돼서 표시

print("첫 번째 줄", end=" ")
print("두 번째 줄")

출력: 첫 번째 줄 두 번째 줄

#### 2. 공백 없이 이어서 출력
- end를 빈 문자열("")로 설정하면, 두 출력이 공백 없이 바로 이어서 출력된다.

print("안녕하세요, ", end="")
print("파이썬!")

출력: 안녕하세요, 파이썬!

#### 3. 다른 구분자 사용
- end를 쉼표와 공백(",")으로 설정하면 각 출력 사이에 쉼표와 공백이 삽입된다.

print("사과", end=", ")
print("바나나", end=", ")
print("체리")

출력: 사과, 바나나, 체리

#### sep = 파라미터
: 기본값이 공백('')으로, 각 인수에 공백이 삽입되어 출력됨
print()에서 인수를 출력할 때, 인수들 사이에 삽입될 문자열을 지정하는 옵션

```python
ex)
print("사과", "바나나", "사과", sep=", ")
출력 : 사과, 바나나, 체리

print("2024", ""04", "27", sep="-")
출력 : 2024-04-27
```

### end = "" : 출력 후 공백을 추가하여 다음 print() 출력 결과가 같은 줄에서 이어지도록 함

또 다른 예로는

#### input().split()을 이용한 입력 처리

##### 1. 한 줄에 입력 받아 변수에 담기 

- input().split() 을 사용하면 사용자가 입력한 문자열을 공백을 기준으로 분리하여 받아올 수 있다.
- 입력받은 것들이 공백으로 구분되어 순서대로 변수에 담기게 된다.
- 나열한 변수의 개수와 입력으로 들어오는 데이터의 개수가 동일해야 한다.
- 모든 데이터는 문자열로 저장된다. 

```python
ex) 
a, b, c = input().split()
print(a)
print(b)
print(c)

입력 : apple banana cherry
출력 : apple
banana
cherry
```


##### 2. 입력받은 데이터의 형 변환

- input().split()으로 분리된 데이터는 모두 문자열 형태로 저장된다. 따라서 필요할 경우 적절한 자료향으로 변환해야 한다.
- map (자료형, input().split()) 을 이용하면 입력받은 데이터 타입을 정할 수 있다. 

ex)
num1, num2 = map(int, input().split())
print("두 수의 합은", num1 + num2, "입니다.")

입력 : 10 20
출력 : 두 수의 합은 30 입니다.  

=> map(int,input().split()) 은 input().split()으로 분리된 문자열들을 int 함수로 변환


#### input().split() 의 작동 과정 이해하기
이전까지는 input().split() 을 기계적으로 사용하여 사용자로부터 여러 값을 입력받고 분할하는 방법을 사용했습니다.
이제 split() 메서드의 작동 원리를 자세히 이해함으로써, input().split() 이 어떻게 작동하는지 더 명확하게 이해할 수 있습니다.

- step : 수 간의 간격, 기본값이 1

### [ continue 와 break 문 ]

#### continue 문

: 현재 반복을 건너뛰고 다음 반복으로 넘어가게 하는 제어문입니다.

다음과 같은 코드에서, num이 짝수인 경우만 반복을 넘기고 싶다면, 다음과 같이 continue문을 쓸 수 있습니다.
```python
for num in range(1, 6):
    print(f"{num}번째 반복문 시작")
    if num % 2 == 0:
        continue
    print(num)

# 출력:
# 1번째 반복문 시작
# 1
# 2번째 반복문 시작
# 3번째 반복문 시작
# 3
# 4번째 반복문 시작
# 5번째 반복문 시작
# 5

```


### 리스트 컴프리헨션이란?

: 반복문(for)와 간단한 표현식을 사용하여, 기존의 list나 반복가능한 객체로부터 새로운 리스트를 짧고 쉽게 만드는 방법이다. 더 짧은 코드로 가독성을 높일 수 있다.

#### 1. 리스트 컴프리헨션의 구조  

새로운\_리스트 = [표현식 for 변수 in 시퀀스]

##### (1) 시퀀스

list :[1,2,3]
range() 함수 : range(3)은 0,1,2를 생성한다.

##### (2) for 변수 in 시퀀스
: "시퀀스에서 하나씩 꺼내서 변수에 담아"라는 의미로, 일반적으로 사용하는 for문과 유사하다.

**for x in [1,2,3] => 리스트 [1,2,3]의 각 숫자를 하나씩 x에 넣게 된다**  
1번째 반복 : x=1
2번째 반복 : x=2
3번쨰 반복 : x=3

##### (3) 표현식

표현식이 x*2면 x에 2를 곱한값을 넣게 되고, x가 1이면 1*2 = 2가 리스트에 추가된다.

ex) 기존 리스트의 숫자에 2를 곱한 new list

arr = [1, 2, 3, 4]
new_arr = [x * 2 for x in arr]
print(new_arr) # 출력: [2, 4, 6, 8]

1. for x in arr 은 arr의 각 숫자를 x에 하나씩 대입
2. x\*2는 x에 2를 곱한 값을 new_arr에 추가함
3. [2,4,6,8] 이라는 새로운 list가 만들어짐

##### for문을 사용한 경우
```python
numbers = [1, 2, 3, 4, 5]
squared = []
for num in numbers:
squared.append(num \*\* 2)
print(squared) # 출력: [1, 4, 9, 16, 25]
```
##### list comprehension을 사용한 경우
```python
numbers = [1, 2, 3, 4, 5]
squared = [num ** 2 for num in numbers]
print(squared) # 출력: [1, 4, 9, 16, 25]
```

books = ['IT책', '역사책', '요리책'] 이라고 가정해볼게요.

**books[1:]**을 하면? → ['역사책', '요리책'] 만 남습니다.

또 거기서 **[1:]**을 하면? → ['요리책'] 만 남습니다.

마지막으로 **[1:]**을 하면? → [](빈 리스트)가 됩니다.


#### 1. books[:1] : "앞에서 딱 1개만!"

콜론(:)이 뒤에 있으면 **"처음부터 ~까지"**라는 뜻입니다.

해석: 인덱스 0번부터 1번 직전까지 가져와!

결과: 첫 번째 항목만 담긴 리스트가 나옵니다.

예: ['IT책', '역사책', '요리책'][:1] → ['IT책']



#### 2. books[:-1] : "마지막 놈만 빼고 다!"

마이너스(-)는 뒤에서부터 숫자를 셉니다. -1은 맨 마지막 놈을 뜻하죠.

해석: 처음부터 뒤에서 첫 번째 놈(-1) 직전까지만 잘라와!

결과: 맨 마지막 항목 하나만 쏙 빠진 리스트가 나옵니다.

예: ['IT책', '역사책', '요리책'][:-1] → ['IT책', '역사책']



#### 3. books[-1:] : "마지막 놈부터 끝까지!" (당연히 가능!)

이것도 아주 많이 쓰는 표현이에요.

해석: 뒤에서 첫 번째 놈(-1)부터 시작해서 끝(:)까지 다 가져와!

결과: 맨 마지막 항목만 담긴 리스트가 나옵니다.

예: ['IT책', '역사책', '요리책'][-1:] → ['요리책']


### [1:]  0번 빼고 두 번째부터 끝까지  ['역사책', '요리책']
### [:1]  시작부터 첫 번째 놈까지 ['IT책']
### [:-1] 처음부터 마지막 놈만 빼고 ['IT책', '역사책']
### [-1:] 마지막 놈만 리스트로 ['요리책']

lower() : 문자열의 모든 문자를 소문자로 변환합니다.

#### strip(), lstrip(), rstrip() : 공백문자 제거

##### strip() : 문자열 양쪽에 있는 공백문자를 제거합니다. 제거되는 문자로는 공백문자( )뿐만 아니라 줄바꿈 문자 \n, 탭 문자 \t 도 제거됩니다.**
```python
text = " Hello, World! "
print(text.strip()) # 출력: "Hello, World!"
```
##### lstrip() : 문자열의 왼쪽(앞쪽)에서 공백과 지정한 문자를 제거합니다.**
```python
text = " Hello, World! "
print(text.lstrip()) # 출력: "Hello, World! "
```
##### rstrip() : 문자열의 오른쪽(뒤쪽)에서 공백과 지정한 문자를 제거합니다.**
```python
text = " Hello, World! "
print(text.rstrip()) # 출력: " Hello, World!"
```
참고: strip() 은 양쪽 공백을 모두 제거하지만, 경우에 따라 왼쪽 또는 오른쪽 공백만 제거하는 것이 더 유용할 수 있습니다. 이럴 때 lstrip() 또는 rstrip() 을 사용하는 것이 좋습니다.

#### 슬라이싱의 기본 문법

슬라이싱의 기본 문법은 다음과 같습니다.

문자열[start:end]

start : 슬라이싱을 시작할 위치의 인덱스입니다. 이 인덱스는 포함됩니다. end : 슬라이싱을 끝낼 위치의 인덱스입니다. 이 인덱스는 포함되지 않습니다. 즉, end 바로 전에 있는 요소까지만 추출됩니다.
```python
# ex)
text = "Python is fun" print(text[1:4]) # 출력: yth
```
#### start 혹은 end 생략하기  

슬라이싱에서는 start 나 end 를 생략할 수 있습니다.

##### start 생략  
: 문자열의 첫 번째 요소(인덱스 0)부터 시작합니다.   

##### end 생략: 문자열의 마지막 요소까지를 의미합니다. 둘 모두 생략할 경우 문자열의 처음부터 끝까지 전체를 의미합니다.
```python
text = "Python is fun" 

print(text[:5]) # 출력: Pytho 
print(text[2:]) # 출력: thon is fun 
print(text[:]) # 출력: Python is fun
```
또한, 음수 인덱스를 사용하여 문자열의 끝에서부터 슬라이싱 범위를 지정할 수도 있습니다.
```python
text = "Python is fun" print(text[-5:]) # 출력: s fun
```
##### 간격(step) 지정하기  

슬라이싱 문법에서는 start , end 외에도 step (간격)을 지정할 수 있습니다. 기본적으로 step 의 값은 1이지만, 2 이상을 주면 해당 간격만큼 건너뛰며 추출합니다.

```python
text = "Python is fun" print(text[::2]) # 출력: Pto sfn

또한 step 에 음수 값을 주면 리스트를 거꾸로 뒤집어서 추출할 수 있습니다.

text = "Python is fun" print(text[::-1]) # 출력: nuf si nohtyP
```

##### 2. 범위(range)의 오해:

문제는 입력받은 그 10개의 숫자 각각이 3이나 5의 배수인지 확인하라는 것입니다.

##### 3. 카운팅 로직:
3의 배수 개수와 5의 배수 개수를 따로 구해서 공백을 두고 출력해야 합니다.
