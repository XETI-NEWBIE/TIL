### 26045. 부분 수열 판별

````python

T = int(input())

for test_case in range(1, T + 1):
    # N: 벨트 길이, M: 타겟 단어 길이
    N, M = map(int, input().split())
    
    # A: 컨베이어 벨트 문자열, B: 만들어야 할 단어
    A = input().strip()
    B = input().strip()
    
    target = 0  # 우리가 B에서 찾아야 할 글자의 인덱스
    result = -1 # 결과값 (못 찾으면 -1)

    # 벨트 A를 처음부터 끝까지 한 번만 훑습니다 (순서대로니까요)
    for i in range(N):
        # 벨트의 글자(A[i])가 우리가 찾는 단어의 현재 글자(B[target_idx])와 같다면?
        # target: "내가 지금 B에서 찾아야 할 글자의 순서" 
        # A[i] : "문자열 A에서 i번째 위치에 있는 글자를 가져와라"라는 뜻
        if A[i] == B[target]:
            target += 1 # 다음 글자를 찾으러 갑니다
            
            # B의 모든 글자를 다 찾았다면?
            if target == M:
                result = i # 현재 벨트의 위치(i)가 정답
                break # 탐색 종료

    print(f'#{test_case} {result}')
````

보니깐 result 처럼 명시적으로 인덱스 / 숫자이 출력되어야 하는 애들은 기본값으로 -1, 1 등을 제대로 설정해놓는 경우가 많은 것 같다. 
result=-1 이라던지 result=1이라던지...
그리고 나중에 if문 가서 바꾸는 것 같다.... 그런데 맨 위의 for 문 변수로 변하는 경우가 많은듯... 위 코드에서 result=i가 된 것처럼..

### Q. B[target]은 B[0], B[1]... 이 맞나요?

### 왜 target == M 이면 끝인가요? 

```PYTHON
  if target == M:
        result = i 
```

여기서 M은 단어의 총 길이입니다. (예: SSAFY는 5글자이므로 $M=5$)

target은 0부터 시작해서 글자를 하나 찾을 때마다 +1씩 늘어납니다.

S를 찾으면 target은 1
두 번째 S를 찾으면 target은 2... 
마지막 5번째 글자인 Y를 찾으면 target은 5가 됩니다.

즉, <b><i>target이 M(단어 길이)과 같아졌다는 건 "단어의 모든 글자를 순서대로 다 찾았다!"</i></b> 는 뜻입니다. 그래서 더 볼 필요 없이 종료(break)하는 거예요.

### Q. result = i 가 왜 "가장 빠른 칸"인가요?

이 부분이 질문자님이 가장 헷갈려 하시는 부분 같은데, 아주 단순한 원리입니다!

벨트는 0번 칸부터 순서대로 흘러갑니다. (for i in range(N))

우리는 벨트가 지나갈 때마다 내가 필요한 글자인지 확인하고, 맞으면 바로 낚아챕니다.

<b><i>마지막 글자를 낚아채는 그 순간의 벨트 위치가 바로 i</i></b> 입니다.

벨트 앞에서부터 가장 먼저 완성되는 시점을 찾는 것이므로, <b><i>처음으로 target == M이 되는 순간의 i 값이 문제에서 요구하는 "가장 빠른 칸의 번호"</i></b> 가 되는 것입니다.

if) SSAFY를 만들 때, 마지막 Y를 8번 칸에서 딱 집었다면?

그때의 i는 8이겠죠? 그래서 result = 8이 되고, "8번 칸에서 단어가 완성됐어!"라고 말하고 8을 출력하게 되는 것입니다.


#### Q. A[i] == B[target]은 무슨 원리인가요?

이 부분은 "**내가 지금 벨트에서 보고 있는 글자가 내가 만들어야 할 단어의 몇 번째 글자인가?**"를 체크하는 곳입니다.

A (벨트): S S F A S F F S Y S

B (목표 단어): S S A F Y

target: "내가 지금 B에서 찾아야 할 글자의 순서" (처음엔 0번인 S를 찾아야 하죠)

진행 과정을 상상해 보세요:

i = 0일 때)  A[0]은 'S'입니다. B[target]도 'S'죠? (일치!)

target을 1로 올립니다. "이제 다음 글자인 B[1]('S')을 찾을 차례야!"라고 표시하는 거죠.

i = 1일 때) A[1]은 'S'입니다. B[1]도 'S'죠? (일치!)

target을 2로 올립니다. "자, 이제 세 번째 글자 B[2]('A')를 찾자!"

i = 2일 때) A[2]는 'F'입니다. 근데 우리는 지금 'A'를 찾아야 해요. (불일치)

그냥 지나갑니다. target은 그대로 2입니다.

이렇게 target은 ***다음에 내가 낚아채야 할 단어의 위치*** 를 가리키는 화살표라고 생각하시면 됩니다.
