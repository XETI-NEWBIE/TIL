## 출력 형식

### 1. 변수 포맷 (%d, %s, ...)과 %를 사용

```python
a = 5
print("A is %d" % a)

b = "apple"
print("B is %s" % b)

print("A is %d and B is %s" % (a, b))
```
맨 뒤에는 % 뒤에 변수를 나열하는 식으로 포맷을 맞출 수 있다. 2개 이상의 변수를 한 문자열에 넣기 위해서는, 소괄호 ()로 감싸서 순서대로 변수들을 나열하면 된다. 

```python
# 출력결과
A is 5
B is apple
A is 5 and B is apple
```
### 2. format 함수를 이용

format 함수를 이용하면 직접 변수의 type을 명시하지 않더라도, 순서 혹은 일므을 명시하여 원하는 변수를 포맷에 맞춰 넣어줄 수 있다. 숫자 대신 새로운 이름 붙여적기도 가능 => 문자열 내 변수에 꼭 {} 붙여줘야 format 이용 가능

```python
a, b = 5, "apple"

print("A is {0}".format(a))
print("A is {new_a}".format(new_a=a))

print("B is {0}".format(b))
print("B is {new_b}".format(new_b=b))

print("A is {0} and B is {1}".format(a, b))
print("A is {new_a} and B is {new_b}".format(new_a=a, new_b=b))
print("B is {1} and A is {0}".format(a, b))
print("B is {new_b} and A is {new_a}".format(new_a=a, new_b=b))
```
```python
A is 5
A is 5
B is apple
B is apple
A is 5 and B is apple
A is 5 and B is apple
B is apple and A is 5
B is apple and A is 5
```

### 3. f - 문자열 포맷을 이용
변수 이름을 중괄호 {}로 감싸면 원하는 변수를 원하는 위치에 딱! 넣어줄 수 있다. 

```python
a, b = 5, "apple"

print(f"A is {a}")
print(f"B is {b}")
print(f"A is {a} and B is {b}")

# 출력결과
A is 5
B is apple
A is 5 and B is apple
```


## 소수점 맞춰 출력하기

```python
a = 33.567268
print("%.4f" % a)

# 출력결과 : 33.5673
```

#### format 함수를 이용하는 것도 가능
기존 포맷이 {0} 이었다면 그 뒤에 :를 붙여 포맷을 지정해주면 된다.

```python
a = 33.567268
print("{0:.4f}".format(a))

# 출력결과 : 33.5663
```
