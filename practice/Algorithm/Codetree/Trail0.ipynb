````python
# 약수가 세 개인 수 (개헷갈리네진짜)

a, b = list(map(int, input().split()))
total_count = 0

for num in range(a,b+1):
    # 언제나 그렇듯 if문 조건 설정 문제
    # int != iterable => 숫자는 for문에 넣어서 하나씩 꺼낼 수 없음
    # Zerodivisonerror : range는 0부터 시작하기 때문에 0으로 나눌라다가 에러뜬거임
    # num%i ====> 1, num+1 로 수정
    # 해당 숫자(num)의 약수가 몇 개인지 셀 변수를 설정해줘야 한다 (숫자가 바뀔 때마다 0으로 초기화) 
    divisor_count = 0
    
    for i in range(1, num+1):
        if num%i==0:
            divisor_count += 1 # num/i로 딱 나누어 떨어지면 약수 개수 +1 
        
    if divisor_count == 3:
        total_count+=1
# 들여쓰기 중요
print(total_count)
````

##### 2중 for문의 계산 과정 (덧셈)

이제 i(행)와 j(열)이 돌면서 각 위치의 값을 더합니다.

i = 0 (첫 번째 줄)

j=0: array1[0][0] + array2[0][0] → 1 + 1 = 2

j=1: array1[0][1] + array2[0][1] → 2 + 1 = 3

j=2: array1[0][2] + array2[0][2] → 3 + 1 = 4

i = 1 (두 번째 줄)

j=0: array1[1][0] + array2[1][0] → 4 + 1 = 5

j=1: array1[1][1] + array2[1][1] → 5 + 1 = 6

j=2: array1[1][2] + array2[1][2] → 6 + 1 = 7

````python
# 새로운 배열 생성
# 원본 배열을 2배로 복사하기 (모든 요소를 2배로 복사)

matrix = [[1,-1],[0,5]]
rows = 2
cols = 2

new_matrix = []
for i in range(rows):
    new_row = []
    for j in range(cols):
        # 각 요소에 2를 곱하여 새로운 행에 추가
        new_row.append(matrix[i][j] * 2)
    new_matrix.append(new_row)
    
print("새롭게 생성된 배열(new_matrix):")
for row in new_matrix:
    print(row)
    
# 출력
"새롭게 생성된 배열(new_matrix):"
[2, -2]
[0, 10]
````

````python
# 특정 원소들의 합

# [문제 요약]
# 4x4 크기의 격자에 정수가 주어질 때, 특정 영역(왼쪽 아래 삼각형 형태)에 
# 해당하는 정수들의 합을 구하는 프로그램을 작성

# [풀이]
# 1. 4x4 격자를 2차원 배열로 입력
# 2. 색칠된 영역의 규칙을 파악
#    - 첫 번째 줄(i=0): 1번째 칸(j=0)까지
#    - 두 번째 줄(i=1): 2번째 칸(j=1)까지
#    - 즉, 열의 인덱스(j)가 행의 인덱스(i)보다 작거나 같은(j <= i) 칸들임

# 4x4 격자 입력 받기
grid = [list(map(int, input().split())) for _ in range(4)]

total_sum = 0

# 격자를 순회하며 조건에 맞는 원소만 더하기
for i in range(4): # 행 반복
    for j in range(4): # 열 반복
        # 색칠된 칸의 조건: 열 인덱스가 행 인덱스 이하인 경우
        if j <= i: 
            total_sum += grid[i][j]

print(total_sum)

# [예제 출력 확인]
# 입력값이 예제와 같을 때: 3 + (2+6) + (3+31+2) + (95+5+7+1) = 155
````

````python

########################################################################################################

# 두 개의 격자 비교하기

# [문제]
# N x M 크기의 격자 두 개가 주어질 때, 같은 위치의 값이 같으면 0, 다르면 1을 출력

# [풀이 ]
# 1. N x M 크기의 격자 2개를 각각 입력받습니다.
# 2. 2중 for문을 돌며 같은 (i, j) 위치의 값을 비교하여 새로운 결과를 출력

import sys

# 행(n), 열(m) 입력 받기
n, m = map(int, input().split())

# 첫 번째 격자 입력
grid1 = [list(map(int, input().split())) for _ in range(n)]

# 두 번째 격자 입력
grid2 = [list(map(int, input().split())) for _ in range(n)]

# 비교 결과 출력
for i in range(n):
    for j in range(m):
        if grid1[i][j] == grid2[i][j]:
            print(0, end=" ")
        else:
            print(1, end=" ")
    print()

########################################################################################################

# 0 나올 때까지 출력

# [문제 요약]
# 정수를 계속 입력받다가 0이 나오면 그 전까지만 출력합니다.

while True:
    try:
        val = int(input())
        if val == 0:
            break
        print(val)
    except EOFError:
        break
````
