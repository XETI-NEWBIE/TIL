Q. Ïôú ÌïòÌïÑ 10Ïù∏Í∞ÄÏöî?  
A. ÏßÄÍ∏à ÌíÄÍ≥† Í≥ÑÏã† Ïà´Ïûê Ïπ¥Îìú Î¨∏Ï†úÏùò Ïà´ÏûêÍ∞Ä 0Î∂ÄÌÑ∞ 9ÍπåÏßÄÏûñÏïÑÏöî?  
Í∞Å Ïà´ÏûêÍ∞Ä Î™á Î≤à ÎÇòÏôîÎäîÏßÄ Ï†ÅÏñ¥Îëò ***'Îπà Î∞©' ÌòπÏùÄ 'Í∏∞Î°ùÏû•'***Ïù¥ 10Ïπ∏ ÌïÑÏöîÌïòÍ∏∞ ÎïåÎ¨∏Ïóê 10ÏùÑ Í≥±Ìï¥ÏÑú ÎØ∏Î¶¨ Î∞©ÏùÑ ÎßåÎì§Ïñ¥ÎëêÎäî Í±∞ÏòàÏöî.

````python
import sys
sys.stdin = open("numbercard.txt", "r")

T=int(input())
for test_case in range(1,T+1):

    n=int(input())
    arr=list(map(int,input()))
    # Î∞©ÏùÄ Ïó¨Ïú†ÏûàÍ≤å Ï§òÏïºÏßÄ...
    count=[0]*10
    # arrÏùò Í∞úÏàòÎßåÌÅº ÎèåÍ≤åÌï¥Ï§òÏïºÌïúÎã§! len(arr)ÏùÑ Ïç®Ï£ºÎ©¥ Îê®
    # for i in range(n) Ïù¥ÎùºÍ≥† Ïç®Ï§òÎèÑ ÎêúÎã§ !!!!!!!!!!
    # for num in arr:
    #    count[num] += 1  Ïù¥Îü∞ Î∞©ÏãùÎèÑ Í∞ÄÎä•ÌïòÎã§!
    for i in range(len(arr)):
        # ÎØ∏ÏπòÍ≤†Îã§ count[i]Í∞Ä ÏïÑÎãàÎùº count[arr[i]]ÏòÄÎã§ Î≠ò Ïñ¥ÎîîÍπåÏßÄ Îñ†Îß•Ïó¨Ï§òÏïºÎåÄ Í∞úÌó∑Í∞àÎ¶¨ÎÑ§
        count[arr[i]]+=1

    max_num=0
    max_count=0

    for card in range(len(count)):
        if count[card] >= max_count:
            max_count=count[card]
            # ÏôÄ Î†àÏ†ÑÎìú Ïù¥Í≤å ÏµúÎåÄÏπòÎã§ Î≠òÌï¥ÏïºÌï†ÏßÄ ÎèÑÏ†ÄÌûà ÏÉùÍ∞ÅÏù¥ ÎÇòÏßÄÏïäÎäîÎã§
            max_num=card
    print(f'#{test_case} {max_num} {max_count}')
````
