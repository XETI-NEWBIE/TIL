## 4836. 색칠하기 문제

```python
T = int(input())

for test_case in range(1, T + 1):
    N = int(input())
    
    # 1. 10x10 텅 빈 도화지(격자) 생성 (0으로 초기화)
    grid = [[0] * 10 for _ in range(10)]
    
    # 2. N번 동안 색칠 작업 진행
    for _ in range(N):
        r1, c1, r2, c2, color = map(int, input().split())
        
        # 주어진 왼쪽 위(r1, c1)부터 오른쪽 아래(r2, c2)까지 반복
        for i in range(r1, r2 + 1):
            for j in range(c1, c2 + 1):
                # 3. 해당 칸에 색상 값을 더해줌
                grid[i][j] += color
                
    # 4. 색칠이 다 끝난 후, 보라색(값이 3인 칸) 개수 세기
    purple_count = 0
    for i in range(10):
        for j in range(10):
            if grid[i][j] == 3:
                purple_count += 1
                
    # 결과 출력
    print(f"#{test_case} {purple_count}")
```

### Q. grid[i][j]는 무엇이고, += color는 왜 하는 걸까?

맞아, grid[i][j]는 (i행, j열)이라는 주소에 있는 칸 그 자체야. 엑셀 시트에서 특정 셀 하나를 딱 찍는다고 생각하면 돼.

근데 왜 그냥 =가 아니라 +=를 쓸까?

grid[i][j] = color라고 하면: "여기에 이 색을 칠해!" (기존 색은 무시하고 덮어씌움)

grid[i][j] += color라고 하면: "여기에 원래 있던 값에 이 색깔 번호를 더해!"

이 '더하기'가 바로 보라색을 찾는 핵심이야.

#### Q. for _ in range(n)은 왜 있고, 왜 그 안에 코드를 써?

여기서 <b><i>"n = 우리가 칠해야 할 사각형의 개수"</i></b> 야. 만약 n이 3이면, 사각형을 3번 칠해야겠지?

1 번째 (빨간색)사각형 정보 읽기 (2 2 4 4 1) → 색칠하기

2 번째 (파란색)사각형 정보 읽기 (3 3 6 6 2) → 색칠하기

3 번째 (빨간/파란색..)사각형 정보 읽기 (...) → 색칠하기

이렇게 "정보 읽기 + 색칠하기" 한 세트를 총 n번 반복해야 하니까, 정보 읽는 코드(input().split())랑 색칠하는 코드(for i... for j...)를 몽땅 n번 반복하는 루프 안에 넣어주는 거야.

비유하자면:
for _ in range(n): 은 "야, 이제부터 사각형 n개 가져올 테니까 하나씩 꺼낼 때마다 바로바로 도화지에 칠해!" 라고 명령하는 거야.

### Q. for i와 for j 중첩 루프는 어떻게 돌아가는 거야?

이게 네가 제일 헷갈려 하는 부분인 것 같아! 코드가 "안쪽에서 바깥쪽으로 도는 것"처럼 느껴질 수 있는데, 정확히는 <b><i>"바깥쪽이 한 걸음 갈 때, 안쪽은 한 바퀴를 다 돈다"</i></b>고 생각해야 해.

```Python
for i in range(r1, r2 + 1):    # i는 '행(Row)' (세로 위치)
    for j in range(c1, c2 + 1): # j는 '열(Column)' (가로 위치)
        grid[i][j] += color
```
바깥 루프 (i): "자, 이제 i행(줄)을 칠할 차례야."

안쪽 루프 (j): "그 줄 안에서 j열(칸)을 하나씩 옆으로 이동하면서 칠할게!"

🚶 실시간 동작 시뮬레이션 (2x2 사각형을 칠한다면?)
i가 첫 번째 줄(r1)에 딱 멈춰 서.

그 상태에서 j가 옆으로 c1부터 c2까지 슥~ 지나가며 칸마다 색을 칠해. (가로 한 줄 완성!)

j가 끝까지 가면 다시 i가 다음 줄(r1 + 1)로 내려가.

다시 j가 그 줄을 왼쪽에서 오른쪽으로 슥~ 칠해.

즉, 행(i) 하나를 고정해두고, 그 줄의 모든 칸(j)을 다 칠한 뒤에 다음 행으로 넘어가는 방식이야. 프린터가 종이를 인쇄할 때 한 줄 쫙 뽑고 종이 슥 올리고, 또 한 줄 쫙 뽑는 거랑 똑같아!

## Q. 왜 for문을 두 번 중첩해서 돌릴까?

사각형은 <b><i>면</i></b>이지? 점이 아니라 수많은 칸(세포)들이 모여서 만들어진 거야. 그래서 이 사각형의 영역 안에 있는 모든 칸을 하나하나 방문해서 색칠해줘야 해.

### 1. for i in range(r1, r2 + 1): (세로 줄 훑기)

"자, 이제 사각형의 <b><i>맨 윗줄(r1)</i></b>부터 <b><i>맨 아랫줄(r2)</i></b> 까지 한 줄씩 내려가면서 작업할 거야!"라는 뜻이야.

### 2. for j in range(c1, c2 + 1): (가로 칸 훑기)

"그 한 줄(i번째 줄) 안에서, <b><i>왼쪽 끝(c1)</i></b>부터 <b><i>오른쪽 끝(c2)</i></b>까지 옆으로 이동하며 칸마다 색을 칠할 거야!"라는 뜻이지.

##  range(10) 대신 쓸 수 있는 더 세련된 방법 (len 사용)

하드코딩(숫자 10을 직접 쓰는 것)보다 더 권장되는 정석적인 표현은 len() 함수를 사용하는 거야. 이렇게 쓰면 격자 크기가 20x20이나 100x100으로 바뀌어도 코드를 수정할 필요가 없거든.

```Python
purple_count = 0
# grid의 행(세로) 길이만큼 반복
for i in range(len(grid)):
    # i번째 행의 열(가로) 길이만큼 반복
    for j in range(len(grid[i])):
        if grid[i][j] == 3:
            purple_count += 1
```
#### <i>len(grid): 리스트의 바깥쪽 길이(행의 개수 = 10)를 가져와!</i>

#### <i>len(grid[i]): 그 안쪽 리스트의 길이(열의 개수 = 10)를 가져와!</i>

📦 2차원 리스트의 구조: 아파트와 방

2차원 리스트는 <b><i>"리스트라는 큰 상자 안에, 또 다른 리스트(작은 상자)들이 들어있는 형태"</i><b>야.

len(grid): 큰 상자 안에 들어있는 작은 상자의 개수 (층수 / Row)

len(grid[i]): i번째 작은 상자 안에 들어있는 알맹이의 개수 (호수 / Column)

"i가 지나온 범위만큼 j가 도는 건가?" (NO!) 아니야! 

j는 i가 한 걸음 갈 때마다 매번 처음부터 끝까지 새로 다 돌아.

<b>i = 0일 때:</b> j는 0번 줄의 처음부터 끝까지(len(grid[0])) 다 돌아.

<b>i = 1일 때:</b> j는 다시 1번 줄의 처음부터 끝까지(len(grid[1])) 새로 다 돌아.

즉, j의 범위가 i에 의해 점점 늘어나는 게 아니라, <b><i>"지금 내가 서 있는 그 줄(i번 줄)의 실제 칸수"</i><b>만큼 도는 거야.

왜 굳이 len(grid[i])라고 정석적으로 쓸까?
네 말대로 $10 \times 10$이면 둘 다 10이니까 상관없어 보여. 하지만 프로그래밍의 정석은 항상 <b>"만약의 상황(변형)"</b> 을 대비하는 거야.

🚫 만약 사각형이 정사각형이 아니라면? (예: 5행 20열)

만약 격자가 가로로 긴 직사각형($5 \times 20$)이라고 해보자.

len(grid)는 5야.len(grid[i])는 20이야. 이때 둘 다 len(grid)를 써버리면 어떻게 될까?
```Python
for i in range(len(grid)):      # 5번 반복
for j in range(len(grid)):  # 이것도 5번 반복?! (원래 20번 돌아야 하는데!)
```
가로로 20칸이나 있는데 앞의 5칸만 검사하고 나머지는 버리게 되겠지? 

그래서 <i>"가로 길이는 그 줄(grid[i])의 길이를 직접 재서 쓰자"</i>는 게 개발자들의 약속이야.

##### 코드의 의미 다시 보기

```Python
purple_count = 0
for i in range(len(grid)):      # 1. "모든 줄(Row)을 하나씩 내려가면서"
for j in range(len(grid[i])):# 2. "방금 들어온 그 줄의 모든 칸(Column)을 옆으로 훑으면서"
    if grid[i][j] == 3:      # 3. "이 칸이 보라색인지 확인해!"
        purple_count += 1
```

문제점: 네 코드의 핵심인 grid[i][j] += color가 박살 나버려!

이유: 빨간색(1)을 칠한 곳에 또 빨간색(1)을 칠하면 1 + 1 = 2가 돼버리지? 원래 2는 파란색인데, 컴퓨터는 이걸 파란색으로 착각하게 되는 대참사가 벌어져.

해결법: 더하기(+=) 전에, "이 칸에 이미 이 색깔이 칠해져 있는지" 확인하는 방어막(if)을 쳐야 해.

```Python
for i in range(r1, r2 + 1):
    for j in range(c1, c2 + 1):
        # 만약 지금 칠하려는 색이 1(빨강)인데, 아직 1이 안 칠해져 있다면?
        if color == 1 and grid[i][j] != 1 and grid[i][j] != 3: 
            grid[i][j] += 1
        # 만약 칠하려는 색이 2(파랑)인데, 아직 2가 안 칠해져 있다면?
        elif color == 2 and grid[i][j] != 2 and grid[i][j] != 3:
            grid[i][j] += 2
```

### 🎨 2순위: "보라색 말고 '순수한 빨간색' 영역의 넓이를 구하시오"

보라색(겹친 부분)을 구하는 게 너무 익숙해졌을 때 뒤통수를 치는 문제야.

변형: 색칠이 다 끝난 후, 파란색과 겹치지 않은 오직 빨간색만 있는 칸의 수를 구하라고 해.

해결법: 이건 오히려 더 쉬워! 도화지를 다 칠해놓고, 마지막에 개수를 셀 때 3이 아니라 1을 찾으면 돼.

```Python
# (색칠하는 과정은 똑같음)
red_only_count = 0
for i in range(len(grid)):
    for j in range(len(grid[i])):
        if grid[i][j] == 1:  # 3(보라) 대신 1(순수 빨강)을 찾음!
            red_only_count += 1
```
### 🌈 3순위: 색깔이 3개로 늘어난다 (빨, 파, 노)

변형: 빨강(1), 파랑(2), 노랑(4) 세 가지 색이 주어지고, 세 색깔이 모두 겹친 영역을 찾으라고 해.

핵심 원리: 왜 하필 1, 2, 3이 아니라 1, 2, 4일까? 겹쳤을 때 나오는 숫자가 절대 다른 조합이랑 안 겹치게 하려고 그래!

빨(1) + 파(2) = 3

빨(1) + 노(4) = 5

파(2) + 노(4) = 6

빨(1) + 파(2) + 노(4) = 7

해결법: 마지막에 if grid[i][j] == 7: 인 칸을 세어주면 돼.

### 📏 4순위: 격자의 크기가 10x10이 아니라 동적으로 주어진다

아까 네가 날카롭게 질문했던 len()과 관련된 변형이야.

변형: 첫 줄에 테스트 케이스가 주어지고, 그다음 줄에 도화지의 크기 N과 M (예: 20x30)이 주어져.

해결법: grid = [[0] * 10 for _ in range(10)] 이 부분을 입력받은 변수로 바꿔주기만 하면 돼.

```Python
# N(행)과 M(열)을 입력받았다고 가정
grid = [[0] * M for _ in range(N)] 

# 나중에 도화지 훑을 때 10 대신 len() 사용
for i in range(len(grid)):
    for j in range(len(grid[i])):
        if grid[i][j] == 3:
            purple_count += 1
```


## 색칠하기 변형 (1) : "같은 색깔끼리도 겹칠 수 있다" 는 조건으로 변경

```python
import sys
sys.stdin = open("color.txt", "r")

T=int(input())
for test_case in range(1,T+1):
    
    n=int(input())
    grid=[[0]*10 for _ in range(10)]
    
    for _ in range(n):
        r1,c1,r2,c2,color=list(map(int,input().split()))

        for i in range(r1,r2+1):
            for j in range(c1,c2+1):
                # 1. 가장 쉬운 방법 (색깔없으면 걍 칠해주기 ㅋㅋ)
                if grid[i][j]!=color:
                    grid[i][j]+=color
                # 2. 가장 디테일한 방법(색깔별로 조건 나눠주기)
                if color==1:
                    if grid[i][j]==0 or grid[i][j]==2:
                        grid[i][j]+=1
                elif color==2:
                    if grid[i][j]==0 or grid[i][j]==1:
                        grid[i][j]+=2
                # 3. 1의 방법에 조건 추가해주기
                # grid에 color가 있거나, grid 칸이 3이라면 색칠하지마  
                if grid[i][j]!=color and grid[i][j]!=3:
                    grid[i][j]+=color

    purple_count=0
    for i in range(len(grid)):
        for j in range(len(grid[i])):
            if grid[i][j]==3:
                purple_count+=1
    
    print(f'#{test_case} {purple_count}')

```

#### 1. 가장 쉬운 방법 (색깔없으면 걍 칠해주기 ㅋㅋ)
```python
if grid[i][j]!=color:
    grid[i][j]+=color
```

문제에 <b><i>"같은 색인 영역은 겹치지 않는다"</i></b>라는 문구가 확실히 있을 때.  

즉, 빨강 위에 빨강이 올 리가 없으니, 이 코드는 사실상 "빨강 위에 파랑" 혹은 "파랑 위에 빨강"이 오는 경우만 체크하게 돼.

장점: 코드가 가장 짧고 간결해서 실수할 확률이 적어.

위험성: 만약 문제에 "같은 색도 겹칠 수 있다"는 말이 슬쩍 들어가면, 이 코드는 빨강+빨강+빨강을 보라색(3)으로 인식하는 대참사를 막지 못해.

### Q. "같은 색도 겹칠 수 있다"는 문제에 1번이 가능한가?

여기가 네가 가장 헷갈려 하는 포인트지? 결론부터 말하면 <b><i>"절반만 정답"</i></b>이야.

성공하는 상황: 빨강(1) 위에 빨강(1)이 바로 또 겹칠 때. → 1 != 1이 아니니까 잘 막아줘.

실패하는 상황 (개잡탕 상황): 빨강(1) + 파랑(2)이 만나서 <b><i>보라(3)</i></b>가 된 곳에, 다시 빨강(1)이 올 때.

지금 칸은 3이고, 새로 온 색은 1이지?

3 != 1은 <b><i>참(True)</i></b>이야!

그래서 컴퓨터는 3 + 1을 해서 <b><i>4</i></b>로 만들어버려. (보라색 파괴!)

그래서 내가 아까 <b><i>방법 3 (!= color and != 3)</i></b>이 "개잡탕 상황"까지 막아주는 완벽한 보험이라고 말한 거야.

#### 2. "완전 통제" (복잡한 응용 버전)

```python
#색깔별 조건 분기
if color==1:
    if grid[i][j]==0 or grid[i][j]==2:
        grid[i][j]+=1
elif color==2:
    if grid[i][j]==0 or grid[i][j]==1:
        grid[i][j]+=2
``` 

##### 언제 쓰는가?

색깔마다 규칙이 다를 때. (예: 빨강+파랑은 보라색인데, 빨강+노랑은 주황색이다!)

디버깅이 필요할 때. 내가 지금 빨강을 칠하는지 파랑을 칠하는지 눈으로 확실히 따라가며 코딩하고 싶을 때 써.

##### 장점: 가장 직관적이야. 논리가 꼬일 일이 거의 없어.

##### 단점: 코드가 길어져. 시험장에서는 시간이 생명인데, if문을 계속 쓰다 보면 들여쓰기 실수나 오타가 날 확률이 높아져.

---

#### 3. "안전 최우선" (개잡탕/함정 대비 버전)

```python
if grid[i][j]!=color and grid[i][j]!=3:
    grid[i][j]+=color
```

### Q. 왜 17, 24, 29라는 숫자가 나왔을까?

테스트 케이스 #1 분석:

빨간색 영역: 9칸

파란색 영역: 16칸

보라색(겹친) 영역: 4칸

계산: (빨강만 있는 5칸) + (파랑만 있는 12칸) = 17칸! 🎯

즉, 너는 지금 <b><i>"순수 빨강 + 순수 파랑"</i>></b>의 개수를 구한 거야.

네가 if grid[i][j] != 2: (또는 비슷한 조건)을 썼다면, 특정 색깔이 겹치는 걸 방해해서 보라색(3)이 만들어지는 과정을 꼬아버렸을 확률이 높아.

### 🛡️ != 3이 이 문제(SWEA)에서 필요 없는 이유

맞아, 채원이 네 말이 일리 있어. 원래 SWEA 문제에서는 != 3이 굳이 없어도 돼.

이유: 문제에서 "같은 색은 안 겹친다"고 했지?

빨강(1) 한 번, 파랑(2) 한 번만 딱 지나가니까 값이 0 → 1 → 3 혹은 0 → 2 → 3으로 딱 끝나.

숫자가 3보다 커질 일이 아예 없으니까 보호할 필요가 없는 거지.

### 🚨 하지만 '재시험 응용'에서는 왜 != 3이 생명줄인가?

내일 재시험에서 ***같은 색도 겹칠 수 있다*** 는 조건이 붙는 순간, != 3이 없으면 네 코드는 무너져.

상황: 빨강(1)이 3번 겹치고 파랑(2)이 1번 겹치는 '개잡탕' 칸이 있다고 치자.

!= 3이 없을 때: 1+1+1+2 = 5가 되어버려.

결과: 마지막에 if grid[i][j] == 3을 찾을 때, 이 칸은 분명 보라색인데 숫자가 5라서 카운트에서 탈락해버려. 답이 틀리게 되는 거지.

그래서 != 3은 "한번 보라색(3)이 됐으면, 더 이상 숫자를 더하지 말고 ***그 상태를 박제해라!***"라는 뜻의 보호막이야.

### A. if 조건문의 의미: "정확해!"

grid[i][j] != color: "이 칸에 이미 빨간색(1)이 칠해져 있지 않고" (***같은 색 중복 칠하기 방지***)

grid[i][j] != 3: "이미 보라색(3)이 된 칸도 아니라면" (***보라색 박제 보호***)

이렇게 하면 빨간색 위에 빨간색을 더해서 2가 되거나, 보라색 위에 빨간색을 더해서 4가 되는 불상사를 완벽하게 막을 수 있어.

### len() 사용과 len(grid[i])의 진짜 의미

len()을 사용하는 게 훨씬 안전하고 세련된 방식인 거 맞아. 나중에 격자 크기가 $10 \times 10$이 아니라 $100 \times 100$으로 바뀌어도 코드를 수정할 필요가 없으니까!

그런데 <i><b>len(grid[i])</b></i>에 대한 이해는 살짝만 교정해 줄게.

### len(grid[i]): i번째 줄의 <i><b>가로 길이(칸수)</b></i>를 의미해.

```python
for i in range(len(grid)):
    for j in range(len(grid[i])):
        if grid[i][j]==1:
            red_count+=1
```

##### len(grid): 아파트 전체 층수(행, Row의 개수)를 재는 거야.

##### len(grid[i]): <i><b>"지금 i층에 방이 몇 개 있니?"</b></i>라고 묻는 거야.

i가 지나간 자리를 뜻하는 게 아니라, <i><b>현재 i번째 행(Row)의 가로 길이(열, Column의 개수)</b></i>를 뜻해.

그래서 for j in range(len(grid[i])):라고 쓰면, i번째 줄의 첫 번째 칸(0)부터 마지막 칸까지 빠짐없이 훑으라는 명령이 되는 거지.
